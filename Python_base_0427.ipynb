## 논리연산

```python
# True : 1
# False : 0
# And : 논리곱
# or : 논리합
# not :부정
```

```python
# True and False
# 둘 중 하나라도 False라면 False
# and는 위와 같은 속성 때문에
# 둘 다 만족해야 하는 '조건'에 사용합니다.

# for i in range(101):
#     if i % 3 == 0 and i % 5 ==0:
#         print(i)
    
for i in range(101):
    if i % 3 == 0:
        if i % 5 == 0:
            print(i)
```

```python
# True and False
# 둘 중 하나라도 True라면 True

for i in range(101):
    if i % 3 == 0 or i % 5 ==0:
        print(i)
    
# for i in range(101):
#     if i % 3 == 0:
#         if i % 5 == 0:
#             print(i)
```

```python
# and가 or보다 먼저 실행됨 (and를 중심으로 괄호묶는다고 생각하기)

for i in range(101):
    if i % 3 == 0 and i % 5 == 0 or i % 2 == 0:
        print(i)

# for i in range(101):
#     if (i % 3 == 0 and i % 5 == 0) or i % 2 == 0:
#         print(i)

for i in range(101):
    if i % 3 == 0 or i % 5 == 0 and i % 2 == 0:
        print(i)

# for i in range(101):
#     if i % 3 == 0 or (i % 5 == 0 and i % 2 == 0):
#         print(i)
```

# 아름다운 코드

* generator
* decorator

```python
#데이터 분석에서 쓰일만한 데코레이터
def 숫자로변환(func):
    def new_func(l):
        print(func(list(map(int, l))))
    return new_func

@숫자로변환
def 평균(l):
    return sum(l) / len(l)

평균([1, 2, 3, 4, 5])
평균(['1', '2', '3', 4, 5])
평균(['1', '2', '3', '4', '5'])
```

# 그리디 알고리즘
* 욕심쟁이 알고리즘이라고 불립니다.
* 가장 최선의 선택한 것이 최종적으로도 가장 최선의 선택임

```python
# 10000원을 들고 가서 1200원짜리 아이스크림 먹었을 때 잔돈의 갯수가 가장 적은 경우를 구하라
잔돈 = 8800

오천원갯수 = 잔돈 // 5000
잔돈 = 잔돈 - 오천원갯수 * 5000

천원의갯수 = 잔돈 // 1000
잔돈 = 잔돈 - 천원의갯수 * 1000

오백원갯수 = 잔돈 // 500
잔돈 = 잔돈 - 오백원갯수 * 500

백원갯수 = 잔돈 // 100
잔돈 = 잔돈 - 백원갯수 * 100

print(f'오천원은 {오천원갯수}개, \
      천원은 {천원의갯수}개, \
      오백원은 {오백원갯수}개, \
      백원 {백원갯수}개를 거슬러 드립니다.')
```

---

마틴이 클린코드에서 주장한 원칙 5개(solid)
SRP : 단일 책임 원칙(각 클래스(또는 모듈, 오브젝트)는 하나의 정보만 가지게 만들어야 한다)
OCP : 개방 폐쇠 원칙(확장에는 열려 있으며, 변경에는 닫혀 있어야 한다)
LSP : 리스코프 치환 원칙(인터페이스의 서브타이핑은 인터페이스에 정의된 형태를 최대한 유지)
ISP : 인터페이스 분리 원칙(인터페이스는 최소한으로 유지)
DIP : 의존 관계 역전 원칙(상위 레벨의 모듈이나 인터페이스가 서브 클래스나 타이핑에 영향을 받아서는 안된다)
