* A - 90점 이상

* C - 60점 이상
* D - 40점 이상

중첩 조건문 또는 연산자를 활용하여 if문을 작성해보세요

```python
score = int(input('점수를 입력하세요(0 ~ 100): '))
absent = int(input('결석 횟수를 입력하세요: '))

if score > 100 or score < 0 or absent < 0:
    print('잘못된 값을 입력하였습니다.')
elif absent < 3:
    if score >= 90 :
        print('등급: A')
    elif score >= 80:
        print('등급: B')
    elif score >= 60:
        print('등급: C')
    elif score >= 40:
        print('등급: D')
    elif score < 40: 
        print('등급: F')
else:
    print('등급: F')
```

문제 4. 태어난 연도를 입력받아 생애주기 받기
2023년 6월 28일 부터 만나이로 변경이 되면서 나이에 혼선이 생기고 있는 가운데
정부 연령기준에 맞춰 자신이 어떤 나이대 인지 찾아보자

https://www.gosims.go.kr/hh/hh001/retrieveLfcySearch.do

* 정부기준으로 영유아는 0~5 / 아동은 6~12 / 청소년 13~18 / 청년 19~29 / 중년 30~49 / 장년 50~64 / 노년 65세 이상 / 연령대 무관으로 나눈다.
* 태어난 연도를 입력 받는다.
* 만나이 적용 날짜를 산출한다.
* if문을 통해 어떤 종류의 생애주기 인지를 출력한다.
* 만나이를 계산하고, 만나이 기준 생애주기를 출력하세요

```python
import datetime
from datetime import timedelta, timezone

year = int(input("태어난 연도를 입력하세요: "))
month = int(input("태어난 월을 입력하세요: "))
day = int(input("태어난 일을 입력하세요: "))
birthday = datetime.date(year, month, day)

d_day = datetime.date(2023, 6, 28)

age = d_day.year - birthday.year - ((d_day.month, d_day.day) < (birthday.month, birthday.day))

if age <= 5:
    cycle = '영유아'
elif age <= 12:
    cycle = '아동'
elif age <= 18:
    cycle = '청소년'
elif age <= 29:
    cycle = '청년'
elif age <= 49:
    cycle = '중년'
elif age <= 64:
    cycle = '장년'
else:
    cycle = '노년'

print(f'만나이: {age}, 만나이 기준 생애주기: {cycle}')
```

```python
## continue ##
구매개수_총가격= [
    (3, 15000),
    (5, 25000),
    (1, 5000),
    (8, 40000),
    (0, 0),
    (2, 10000)
]

for 구매개수, 총가격 in 구매개수_총가격:
    if 구매개수 < 1:    
        continue    # 구매개수가 0인 것은 출력X
    print('구매개수는 {}개이며 {}원입니다.'.format(구매개수, 총가격))
```

```python
# 조건문 문제
# 실행결과가 0인 코드를 작성하고 싶을때 어떻게 작성해야하는가?

money = 2000
snack = 500
water = 1000
while money !=0:
  if money > 1000:
    money -= water
  elif 0 < money <= 1000:
    money -= snack * 2
  else:
    print("")
print(money)
  
# 6. 답 없음
```

```python
import random

guess_number = random.randint(1,100)
user_input = int(input("숫자를 입력해서 값을 맞춰보세요."))

while user_input != guess_number:
    user_input = int(input("숫자를 입력해서 값을 맞춰보세요."))
    if user_input > guess_number:
        print(f'입력값: {user_input}, 값이 큽니다.')
    else:
        print(f'입력값: {user_input}, 값이 작습니다.')
    user_input = int(input("숫자를 입력해서 값을 맞춰보세요."))
else:
    print(f'입력값: {user_input}, 답이 맞았습니다.')
```
