## 기본 사용방법

matplotlib.pyplot 모듈은 MATLAB과 비슷하게 명령어 스타일로 동작하는 함수의 모음입니다. <br/>
matplotlib.pyplot 모듈의 각각의 함수를 사용해서 간편하게 그래프를 만들고 변화를 줄 수 있습니다.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4]) # y축만 설정
plt.show()

'''
pyplot.plot() 함수에 하나의 숫자 리스트를 입력함으로써 아래의 그래프를 그린다.
plot()함수는 리스트의 값들이 y값들이라고 가정하고 x값 [0, 1, 2,3]을 자동으로 만들어낸다.
'''
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16]) # x축, y축 설정
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # red, 원형('o') marker
plt.plot([1, 2, 3, 4], [16, 9, 4, 1], 'b-') # blue, 실선('-')
plt.axis([0.8, 6, 0, 20]) # [xmin, xmax, ymin, ymax]
plt.show()
```

```python
# 여러개의 그래프를 그리는 방법 (참고)

import matplotlib.pyplot as plt
import numpy as np

# 200ms 간격으로 균일하게 샘플된 시간
t = np.arange(0., 5., 0.2)

# 빨간 대쉬, 파란 사각형, 녹색 삼각형
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# 레이블이 있는 데이터 사용하기

import matplotlib.pyplot as plt

data_dict = {'data_x' : [1, 2, 3, 4, 5], 'data_y' : [2, 3, 5, 10, 8]}

plt.plot('data_x', 'data_y', data = data_dict)
plt.show()
```

```python
# x축, y축 라벨 붙이기

import matplotlib.pyplot as plt

# 라벨 위치
location_x = 'right'    # 'left', 'center'
location_y = 'top'      # 'bottom', 'center'

# 라벨 폰트
font_x = {'family': 'serif', 'color': 'b', 'weight': 'bold', 'size': 14}
font_y = {'family': 'fantasy', 'color': 'deeppink', 'weight': 'normal', 'size': 20}

plt.plot([1, 2, 3, 4, 5], [2, 3, 5, 10, 8])
plt.xlabel('X-Axis', labelpad = 15, loc = location_x, fontdict = font_x)
plt.ylabel('Y-Axis', labelpad = 20, loc = location_y, fontdict = font_y)
plt.show()
```

```python
# 범례 사용하기

import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4, 5], [2, 3, 5, 10, 8], label = 'Price ($)')    # 범례 설정하기
plt.plot([1, 2, 3, 4, 5], [3, 5, 9, 7, 12], label = 'Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# plt.legend(loc = (1.0, 1.0))    # 범례 활성화
plt.legend(loc = 'best', ncol = 2, fontsize = 14)

plt.show()
```

```python
# 축 범위 얻기 
# xlim(), ylim() 함수는 그래프 영역에 표시되는 X축, Y축의 범위를 각각 반환합니다.
# axis() 함수는 그래프 영역에 표시되는 X, Y축의 범위를 반환합니다.

import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

x_range, y_range = plt.xlim(), plt.ylim()
print(x_range, y_range)

axis_range = plt.axis('scaled')
print(axis_range)

plt.show()
```

```python
# 그래프의 영역을 나타내기 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)

plt.show()
```

## 색상, 마커, 선 종류

```python
# 선 종류 지정하기

import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], '-', color='C0', label='Solid')
plt.plot([1, 2, 3], [3, 3, 3], '--', color='C0', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':', color='C0', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='C0', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='C0', label="'solid'")
plt.plot([1, 2, 3], [3, 3, 3], linestyle='dashed', color='C0', label="'dashed'")
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', color='C0', label="'dotted'")
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', color='C0', label="'dashdot'")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.tight_layout()
plt.show()
```

```python
'''
튜플을 사용해서 선의 종류를 커스터마이즈할 수 있습니다.
예를 들어, (0, (1, 1))은 ‘dotted’와 같고, (0, (5, 5))는 ‘dashed’와 같습니다.
또한 (0, (3, 5, 1, 5))는 ‘dashdotted’와 같습니다.
숫자를 바꿔가면서 다양한 선의 종류를 만들어보세요.
결과는 아래와 같습니다.
'''

import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], linestyle=(0, (1, 1)), color='C0', label='(0, (1, 1))')
plt.plot([1, 2, 3], [3, 3, 3], linestyle=(0, (1, 5)), color='C0', label='(0, (1, 5))')
plt.plot([1, 2, 3], [2, 2, 2], linestyle=(0, (5, 1)), color='C0', label='(0, (5, 1))')
plt.plot([1, 2, 3], [1, 1, 1], linestyle=(0, (3, 5, 1, 5)), color='C0', label='(0, (3, 5, 1, 5))')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
```

```python
# 선 끝 모양 지정하기
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', linewidth=10,
      solid_capstyle='butt', color='C0', label='solid+butt')
plt.plot([1, 2, 3], [3, 3, 3], linestyle='solid', linewidth=10,
      solid_capstyle='round', color='C0', label='solid+round')

plt.plot([1, 2, 3], [2, 2, 2], linestyle='dashed', linewidth=10,
      dash_capstyle='butt', color='C1', label='dashed+butt')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashed', linewidth=10,
      dash_capstyle='round', color='C1', label='dashed+round')


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
```

## 다양한 그래프 그리기

```python
# 1. 막대그래프
# plt.bar(x, y) : 막대그래프 그리기
# color : 색상 지정

import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2]
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values, color = 'y')
plt.xticks(x, years)

plt.show()
```

```python
# 1. 막대그래프
# 값별로 색상 다르게 하기
# width : 그래프 폭 지정

import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2]
years = ['name1', 'name2', 'name3']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.bar(x, values, color = colors, width = 0.4)
plt.xticks(x, years)

plt.show()
```

```python
# 1. 막대그래프
# align = 'edge' : 엣지 표시하기
# color : 색상 지정

import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2]
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values, align='edge', edgecolor='lightgray',
        linewidth=5, tick_label=years)

plt.show()
```

```python
# 1. 막대그래프 - 가로막대
# plt.barh(y, x) : 가로막대 그리기

import matplotlib.pyplot as plt
import numpy as np

y = [0,1,2]
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y, values)
plt.yticks(y, years)

plt.show()
```

```python
# 2. 산점도
# plt.scatter(x, y) : 산점도 그리기

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)

plt.scatter(x, y, c = colors)
plt.show()
```

```python
# 3. 히스토그램
# plt.hist(data) : 히스토그램 그리기

import matplotlib.pyplot as plt

weight = [10, 20, 30, 20, 10, 40, 30, 20]

plt.hist(weight, bins = 10, label = 'bins=10')
plt.hist(weight, bins = 30, label = 'bins=30')
plt.legend()

plt.show()
```

```python
# 4. 파이차트
# plt.pie(ratio, labels, autopct) : 파이차트 그리기

import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels = labels, autopct = "%.1f%%") # autopct : 부책꼴의 표시 숫자형식
plt.show()
```
