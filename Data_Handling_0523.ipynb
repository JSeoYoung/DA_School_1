## **난수 생성 함수**

* **np.random.**

* rand( ) randn( ) randint( ) shuffle( )

* choice( ) permutation( )

* normal( )

```python
a = np.random.rand(2, 2)    # 0 ~ 1 사이에 값 랜덤
print(a)
```

```python
b = np.random.randn(2, 2)   # 평균이 0, 분산이 1 : 표준 정규 분포
print(b)
```

```python
c = np.random.randint(0, 10, size=5)    # 랜덤한 정수값 생성, (최소, 최대, size)
print(c)
```

```python
c2 = np.random.randint(0, 10, size=(2, 2))
print(c2)
```

```python
d = np.arange(10)
np.random.shuffle(d)    # 원본 자체가 바뀜
print(d)
```

```python
e = np.random.choice(d, 5)   # 원하는 갯수만큼 뽑음 (중복 허용)
e_false = np.random.choice(d, 5, replace=False)   # 중복 허용 X
e_percent = np.random.choice(d, 5, p=[0.1, 0.1, 0.1, 0.1, 0.2, 0.1, 0.1, 0.1, 0.1, 0])  # 지정확률, sum = 1
e_false_percent = np.random.choice(d, 5, replace=False, p=[0.1, 0.1, 0.1, 0.1, 0.2, 0.1, 0.1, 0.1, 0.1, 0])
print(f"e: {e}")
print(f"e_false: {e_false}")
print(f"e_percent: {e_percent}")
print(f"e_false_percent: {e_false_percent}")
```

```python
# permutation 순열
f = np.arange(10)
f_p = np.random.permutation(f)  # 원본 데이터 유지
print(f"f: {f}")
print(f"f_p: {f_p}")
```

```python
g = np.random.normal(0, 1, 3)   # 정규분포, (평균, 표준편차, 갯수)
print(g)
```


* <h3>loc, iloc -> 일반 슬라이싱과 다른 점은 DataFrame 가면 더 차이를 느낄 수 있을 듯함</h3>


```python
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print(s)
```

```python
print(s['b'])
print(s.loc['b'])   # 행에 접근, 레이블 사용하여 접근
print(s['b':'d'])
print(s.loc['b':'d'])   # 'd' 포함
```

```python
print(s.iloc[1]) # (행에 기반한) 열에 접근
print(s.iloc[1:3])  # [3] 미포함
```

* <h3>at, iat</h3>
* <h3>get</h3>

```python
print(s.at['a'])    # at : loc 비슷, 하나의 값만 가져올 때
print(s.iat[1])     # at에 기반
```

```python
print(s.get('e', default = -1))   # 없는 index일 경우, 없다는 것을 알려줌
```

**<h3><< 결측값 (NaN, None) >></h3>**

- isna()
- isnull()

**NaN(Not a Number)** -> datatype = float <br/>
**None** -> datatype = none

```python
import numpy as np
import pandas as pd

data = [1, 2, 3, None]

print(np.array(data))   # None으로 인식
print(pd.Series(data))  # NaN으로 인식 -> float형, 오류를 줄일 수 있음
```

```python
s = pd.Series(data)

print(s.isna())
print(s.isnull())
print(s.isnull().sum())
```

```python
print(s.notnull())
print(s.notnull().sum())
```

**<h4>결측치 처리하기</h4>**

```python
s.dropna()  # 제거
```

```python
s.fillna(0) # 치환
```
