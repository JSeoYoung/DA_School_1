# **별이 빛나는 밤**

```python
print("*")
```

```python
for i in range(5):
    print("*")
```

```python
for i in range(5):
    print("*", end = "")
```

```python
for i in range(5):
    for j in range(5):
        print("*", end = "")
    print()
```

```python
for i in range(5):
    print("*" * 5)
```

```python
for i in range(5):
    for j in range(i+1):
        print("*", end = "")
    print()
```

```python
for i in range(5):
    print("*" * (i+1))
```

# **줄 바꿔 출력하기**

```python
x = int(input("숫자를 입력해주세요: "))

for i in range(x):
    print(i+1)
```

```python
x = int(input("숫자를 입력해주세요: "))

for i in range(x,0,-1):
    print(i)
```

```python
x = int(input("숫자를 입력해주세요: "))

for i in range(1, x+1):
    print(f"{i}", end="\t")
    
    if i%10 == 0:
        print()
```

# **오늘의 당첨번호**

```python
import random

number = random.sample(range(1,51),3)
print(number)
```

```python
import random

input_num = int(input("구매할 로또 갯수를 입력해주세요: "))
lotto_num = []

for i in range(input_num):
    lotto_num.extend(random.sample(range(1,46),6))

lotto_num.sort()
print(lotto_num)
```
