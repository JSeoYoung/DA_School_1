```python
# test case : "원범 원범 혜원 혜원 혜원 혜원 유진 유진"

string = input('투표 결과를 입력해주세요: ')

string_list = string.split(' ')
string_set = set(string_list)
count_dict = dict()

for key in string_set:
    count_dict[key] = string_list.count(key)

print(f'{max(count_dict.keys())}이 {max(count_dict.values())}표를 받아서 반장이 되었습니다.')
```

```python
limit_weight = float(input('제한 무게를 입력하세요: '))
number = int(input('친구 수를 입력하세요: '))

total_weight = 0
for i in range(number):
    total_weight += float(input(f'{i+1}번째 친구 몸무게를 입력해주세요: '))
    if total_weight > limit_weight:
        print(f'{i+1}번째 친구부터는 다음 기구에 탑승해주세요.')
        break
else:
    print(f'총 {number}명 탑승하였습니다.')
```

```python
# 법인 택시기사들의 평균 사납금: 기본 13만 5천원
# 매출 30만원을 바람.
# 손님별 운행시간: 5~20분 / 5~15분인 고객의 콜을 받음
# 기본요금 : 기본 5분일 경우 5000원, 1분 추가될때마다 1000원 추가

import random

# [매칭횟수, 탑승손님수, 소요시간, 총수익]
taxi = {'call': 0, 'boarding': 0, 'time': 0, 'money': 0}

while taxi['money'] < 300000:
    taxi['call'] += 1
    boarding_time = random.randint(5,21)
    if 5 <= boarding_time <=15:
        taxi['boarding'] += 1
        taxi['time'] += boarding_time
        taxi['money'] += 5000 + 1000*(boarding_time-5)
        # print('손님{}이(가) 탑승했습니다. 소요시간은 {}분이 걸립니다. 금액은 {}원입니다.'.format(taxi['boarding'], boarding_time, 5000 + 1000*(boarding_time-5)))
        print(f'손님{taxi["boarding"]}이(가) 탑승했습니다. 소요시간은 {boarding_time}분이 걸립니다. 금액은 {5000 + 1000*(boarding_time-5)}원입니다.')
    else:
        print('매칭에 실패했습니다.')
else:
    print('오늘 콜 횟수는 {}번이고, 탑승손님수는 {}번이고, 소요시간은 {}분이고, 총 수익은 {}원입니다.'.format(taxi['call'], taxi['boarding'], taxi['time'], taxi['money']))
```

# **예외처리**



오류가 나면 파이썬은 에러가 나서 수행이 중지가 된다.
1. 중지되지 않고 수행
2. 예외적인 부분이나 에러를 좀더 효과적으로
3. 기타

[오류의 종류]
* 프로그램 실행 전에 발생
* 프로그램 실행 중에 발생

[예시]
* ML 모델 돌리는데 데이터에 Null값이 있을 때
* 크롤링시 URL주소를 잘못 입력했을 때
* 숫자형 대신에 누군가 문자형을 입력했을 때

[예외 종류]
* AttributeError - 존재하지 않는 속성 지정
* IndexError - 존재하지 않는 범위의 인덱스 지정
* KeyError - 존재하지 않는 키를 지정
* TypeError - 올바르지 않는 타입 지정
* ValueError - 올바르지 않는 값 지정
* ZeroDivisionError - 0으로 나눔

## **Try - except 문**

```python
try:
    명령
except:
    예외 처리 명령
```

```python
try:    # 꼭 실행되어야 할 구문
    num = int(input('숫자를 입력해주세요: '))
except: # try가 실행되지 않았을 때 except 실행
    print('숫자 외에 자료형이 입력되었어요')
```

```python
try:
    num = int(input('숫자를 입력해주세요: '))
except ValueError:  # erorr명 추가
    print('숫자 외에 자료형이 입력되었어요')
```

```python
def div_num(a,b):
    '''나눗셈 결과를 print하는 함수'''
    try:
        val = a / b
        print(val)
    except ZeroDivisionError:
        print('Error: 0으로 나누었습니다.')
    except TypeError:
        print('Error: 연산할 수 없는 인수를 지정했습니다.')
    except Exception:
        print('Error: 알 수 없는 예외가 발생했습니다.')

div_num(3,0)
```

## **Try - Except - else 문**

```python
try:    # 예외가 발생할 것 같은 구문
    num = int(input('숫자를 입력해주세요: '))
except: # 예외가 발생했을 때 출력하는 구문
    print('숫자가 입력되지 않았습니다.')
else:   # 예외가 발생하지 않았을 때 문장 실행
    print(f'입력된 숫자는 {num}입니다.')
```

## **Try - Except - else - finally 문**

```python
try: #예외가 발생할것 같은 구문
  num = int(input("숫자를 입력해주세요"))

except: #예외가 발생했을때 출력하는 구문
  print("숫자가 입력되지 않았습니다.")

else: #예외가 발생하지 않았을때 문장 실행
  print(f"입력된 숫자는 {num} 입니다.")

finally: #무조건 실행
  print("이 구문은 무조건 실행됩니다.")
```

```python
def div_num(a, b):
    try:
        val = a / b
        print("나눗셈 결과: {}".format(val))
    except:
        print("예외가 발생했습니다.")
    else:
        print("처리를 정상 종료했습니다.")
    finally:
        print("처리를 종료했습니다.")


print('----- 정상 처리 시 -----')
div_num(4, 2)
print('----- 예외 발생 시 -----')
div_num(10, 0)
```

## **사용자 정의 예외처리하기**

```python
class MyError(Exception): #새로운 오류를 정의하려면 Exception을 상속해야함
  pass

try: #예외가 발생할거같은 구문
  num = int(input("숫자를 입력해주세요"))
except: #예외가 발생했을때
  print("숫자가 입력되지 않았습니다.")
  raise MyError
except MyError:
  print("숫자가 입력되지 않아 MyError 호출 하였습니다.")
else: #예외가 발생되지 않았을때 문장 실행
  print("입력된 숫자는", num, "입니다")
finally:
  print("이 구문은 무조건 실행됩니다.")
```

## 연습문제

```python
class SoldOutError(Exception):
    pass
class MaxOrderError(Exception):
    pass

cake = 4
waiting = 1
i = 0

while True:
    try: 
        print("남은 케이크는", cake, "입니다")
        order = int(input("손님 케이크 몇개 주문하시나요?"))
        if cake == 0:
            raise SoldOutError
        elif order > cake:
            print("손님 남은 케익이", cake,"라서 주문이 불가하세요")
        elif order <= 0:
            raise ValueError
        elif order > 2:
            raise MaxOrderError
        else:
            print("케익", order,"개 포장해주세요")
            waiting += 1
            cake -= order #주문 수 만큼 케이스 감소
            i += order
    except ValueError:
        print('잘못된 값을 입력했습니다.')
    except SoldOutError:
        print('남은 케익이 없습니다. --- Sold Out ---')
        break
    except MaxOrderError:
        print('최대 주문량은 2개 입니다.')
```

## **위치 인수와 키워드 인수**

```python
def billing(unit_prices, units, fees, tax_rate):
  return (unit_prices * units + fees) * tax_rate
```

```python
## 위치 인수 ##
# 위치인수는 호출할때 함수의 정의와 인수 순서가 같은 순서로 지정!!
# 위치가 틀어지면 값이 완전히 바뀔수 있다.

x = billing(1000, 10, 500, 1.1)
print(x)
```

```python
## 키워드 인수 ##

x = billing(
            unit_prices = 1000, 
            units = 10, 
            fees = 500, 
            tax_rate = 1.1
            )
print(x)
```

```python
# 위치 + 키워드
# 위치 인수를 먼저 쓰고, 키워드 인수를 써야 오류가 나지 않는다.

x1 = billing(1000, 10, 500, tax_rate = 1.1)
# x1 = billing(1000, 10, tax_rate = 1.1, 500)   # error
print(x1)
```

```python
def print_max(x, y, z):
    if x > y: 
        if x > z:
            print(f'가장 큰 수: {x}')
        else:
            print(f'가장 큰 수: {z}')
    else:
        if y > z:
            print(f'가장 큰 수: {y}')
        else:
            print(f'가장 큰 수: {z}')

a = float(input('첫번째 수를 입력하세요: '))
b = float(input('두번째 수를 입력하세요: '))
c = float(input('세번째 수를 입력하세요: '))
print_max(a, b, c)
```

```python
def print_even(input_list):
    for a in input_list:
        if a % 2 == 0:
            print(a)

print_even([1,3,2,10,12,11,15])
```
